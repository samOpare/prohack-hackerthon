```python
from zipfile import ZipFile

#unzipping the zipfile
with ZipFile("prohack_dataset.zip") as zipObj:
    fileNames = zipObj.namelist()
    for file in fileNames:
        if file.endswith(".csv"):
            zipObj.extract(file,"data")
    
```

### loading the data into data frames

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_squared_error
import math

train = pd.read_csv("data/train_to_download_final_1.csv")
train.head()
```

```python
test = pd.read_csv("data/test_to_download_final_1.csv")
test.head()
```

Exploratory Data Analysis(EDA)

```python
from matplotlib import pyplot as plt
%matplotlib inline

train.plot(x= "Gross income per capita", y= "y", kind = "scatter")
# plt.show()
```

### Prediction

```python
# Create our model
model = RandomForestRegressor(n_estimators=200)

# Split the data into inputs (X) and desired output (y)
X, y = train[test.columns[2:]].fillna(0), train['y'].fillna(0)

# Fit the model
model.fit(X, y)

#predictions
y_pred = model.predict(X)

# Score the model
math.sqrt(mean_squared_error(y, y_pred))# squared=False tells this to return RMSE
```

```python
# Get the model's predictions for the test set and add them as a column (y) to test
preds = model.predict(test[test.columns[2:]].fillna(0))
test['y'] = preds
test.head()
```

### Submission

```python
ss = pd.DataFrame({
    'Index':test.index,
    'pred':test['y'],
    'opt_pred':0
})
ss.head()
```

```python
ss = pd.DataFrame({
    'Index':test.index,
    'pred':test['y'],
    'opt_pred':0, 
    'eei':test['existence expectancy index'] # So we can split into low and high EEI galaxies
})
# Fix opt_pred
n_low_eei = ss.loc[ss.eei < 0.7].shape[0]
n_high_eei = ss.loc[ss.eei >= 0.7].shape[0]
ss.loc[ss.eei < 0.7, 'opt_pred'] = 99 # 66*99 = 6534 - >10%, <100 each
ss.loc[ss.eei >= 0.7, 'opt_pred'] = (50000-6534)/n_high_eei # The rest to high eei gs
# Leaving 5k zillion whatsits to the admin
ss = ss.drop('eei', axis=1)
ss.to_csv('Sub_test.csv', index=False)
ss.head()
```
